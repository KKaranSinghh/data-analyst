```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
# countplot : survival based on gender                          (palette) is for colour     
sns.countplot(df, x = 'survived',hue='sex',)#palette='RdBu')
plt.title('survivedvs not survived by gender')
plt.show()
```
