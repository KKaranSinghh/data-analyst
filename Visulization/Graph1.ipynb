```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
# countplot : survival based on gender                          (palette) is for colour     
sns.countplot(df, x = 'survived',hue='sex',)#palette='RdBu')
plt.title('survivedvs not survived by gender')
plt.show()
```

```python
sns.barplot(df, x = 'class', y ='fare', hue ='sex')
plt.title('average fare by passenger class and gender')
plt.show
```

```python
sns.violinplot(df, x= 'class', y ='age',hue= 'sex')
    
```

```python
sns.pointplot(df, x= 'class', y= 'survived', hue='sex',palette= 'dark',markers=['o','x'])
plt.title('survival rate by class and gender')
plt.show
```
